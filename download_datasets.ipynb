```python
import os
os.environ['HF_ENDPOINT'] = 'https://hf-mirror.com'

from datasets import load_dataset

def download_and_save():
    # 设置国内镜像站 hf-mirror
    os.environ['HF_ENDPOINT'] = 'https://hf-mirror.com'
    
    dataset_name = "Salesforce/wikitext"
    dataset_config = "wikitext-2-raw-v1"
    save_directory = "dataset"
    
    print(f"Downloading {dataset_name} ({dataset_config}) from HF mirror {os.environ['HF_ENDPOINT']}...")
    
    # Load the dataset from Hugging Face (using mirror)
    dataset = load_dataset(dataset_name, dataset_config)
    
    # Save the dataset to the specified local directory
    dataset.save_to_disk(save_directory)
    
    print(f"Dataset successfully downloaded and saved to the '{save_directory}' directory.")

if __name__ == "__main__":
    download_and_save()
```
